Draw the Entity Relationship (“ER”) Diagram for the database described below:

This database archives and rents movies to its customers online. The records of customers, movies, and rentals are stored on an information system.

A customer, who must be 18 years or older, creates his record online with ID, name, contact info and birthdate. He can register up to 5 digital devices with unique IDs for his/her own use.

A digital device is either a smart phone, tablet, smart TV or computer and it must belong to a unique customer.

A customer can search movies in the database based on name, director, and cast. When a customer decides to rent one or more movies online, s/he first pays for them via an online payment method (e.g. PayPal, iDeal, credit card, ...).

Generate sql of the designed diagram by using a database tool ( You can use lucidchart: https://www.lucidchart.com/pages ).



```python
sql_script= """
CREATE TABLE Customer (
    CustomerID INT PRIMARY KEY ,
    Name VARCHAR(255) NOT NULL,
    ContactInfo VARCHAR(255),
    Birthdate DATE NOT NULL CHECK (TIMESTAMPDIFF(YEAR, Birthdate, CURDATE()) >= 18)
);

CREATE TABLE Stats (
    DeviceID INT PRIMARY KEY AUTO_INCREMENT,
    DeviceType ENUM('Smartphone', 'Tablet', 'Smart TV', 'Computer') NOT NULL,
    CustomerID INT NOT NULL,
    FOREIGN KEY (CustomerID) REFERENCES Customer(CustomerID) ON DELETE CASCADE
);

CREATE TABLE Movie (
    MovieID INT PRIMARY KEY AUTO_INCREMENT,
    Name VARCHAR(255) NOT NULL,
    Director VARCHAR(255),
    Cast TEXT
);

CREATE TABLE Rental (
    RentalID INT PRIMARY KEY AUTO_INCREMENT,
    CustomerID INT NOT NULL,
    MovieID INT NOT NULL,
    PaymentMethod ENUM('PayPal', 'iDeal', 'Credit Card', 'Other') NOT NULL,
    RentalDate DATETIME DEFAULT CURRENT_TIMESTAMP,
    ExpiryDate DATETIME GENERATED ALWAYS AS (DATE_ADD(RentalDate, INTERVAL 90 DAY)) STORED,
    FOREIGN KEY (CustomerID) REFERENCES Customer(CustomerID) ON DELETE CASCADE,
    FOREIGN KEY (MovieID) REFERENCES Movie(MovieID) ON DELETE CASCADE
);

CREATE TABLE PlayStatistics (
    PlayID INT PRIMARY KEY AUTO_INCREMENT,
    RentalID INT NOT NULL,
    DeviceID INT NOT NULL,
    PlayDateTime DATETIME DEFAULT CURRENT_TIMESTAMP,
    PlayDuration TIME,
    FOREIGN KEY (RentalID) REFERENCES Rental(RentalID) ON DELETE CASCADE,
    FOREIGN KEY (DeviceID) REFERENCES Stats(DeviceID) ON DELETE CASCADE
);
"""

sql_file_path = "/mnt/data/movie_rental_schema.sql"
with open(sql_file_path, "w") as file:
    file.write(sql_script)

# Provide the download link to the user
sql_file_path
```
